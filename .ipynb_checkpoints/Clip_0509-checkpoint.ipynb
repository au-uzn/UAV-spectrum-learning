```python
import numpy as np
from PIL import Image
from matplotlib import pyplot as plt

im = np.array(Image.open('0066/IMG_0066_1.tif'))
print(im.shape)
plt.imshow(im, interpolation='nearest')
plt.show()

im = np.pad(im, (1, 1), 'constant', constant_values=(0,))

size = 160
num = (np.int(im.shape[0]/size), np.int(im.shape[1]/size))
imageSet = []
for i in range(num[0]):
    for j in range(num[1]):
        imageSet.append(im[i*size:(i+1)*size+2, j*size:(j+1)*size+2])
fig , ax = plt.subplots(num[0], num[1], sharex=True, sharey=True)

index = 0
for i in range(num[0]):
    for j in range(num[1]):
        plt.subplot(ax[i,j])
        plt.imshow(imageSet[index])
        index = index + 1
```
